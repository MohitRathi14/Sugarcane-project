```python
# import the opencv l
import cv2 as cv
```

```python
image = cv.imread('C:\\Users\\rajur\\OneDrive\\Desktop\\pythonProject\\Image\\image1.jpeg')
```

```python
# display the image
cv.imshow('Image', image)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
resized_image = cv.resize(image, (500, 500), interpolation=cv.INTER_LINEAR)   
```

```python
# display the image
cv.imshow('Image',resized_image )
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# to convert the image color space from BGR to RGB
rgb_image = cv.cvtColor(resized_image, cv.COLOR_BGR2RGB)
```
